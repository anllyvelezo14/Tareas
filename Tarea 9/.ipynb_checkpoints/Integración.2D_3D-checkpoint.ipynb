```python
from mpl_toolkits.mplot3d import Axes3D  

import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np
```

```python
#Cargando coeficientes
coeff=np.loadtxt('coeff.dat')
#print(coeff)
```

```python
#Función a integrar
def f(x,y):
    value=(1/np.pi)*(np.exp(-x**2-y**2))
    return value

#Limite inferior y superior de la integral interna
def c(x):
    value=-2
    return value

def d(x):
    value=2
    return value

#Limite inferior y superior de la integral externa
a=-2
b=2
```

```python
#Inicia integración
def DoubleGauss(n,m):
    #Creando listas de pesos y raices
    u_i_list=[]
    w_i_list=[]
    t_j_list=[]
    w_j_list=[]
    for i in range(0,coeff.shape[0]):
        if coeff[i][0]==n:
            u_i_list.append(coeff[i][1])
            w_i_list.append(coeff[i][2])            
    for j in range(0,coeff.shape[0]):
        if coeff[j][0]==m:
            t_j_list.append(coeff[j][1])
            w_j_list.append(coeff[j][2])

    sum_i=0
    for i in range(0,n):
        sum_j=0
        x_i=((b-a)*u_i_list[i]+a+b)*0.5
        for j in range(0,m):
            y_j=((d(x_i)-c(x_i))*t_j_list[j]+d(x_i)+c(x_i))*0.5
            sum_j+=w_j_list[j]*f(x_i,y_j)
        sum_i+=w_i_list[i]*(d(x_i)-c(x_i))*sum_j
        
    value=sum_i*(b-a)*0.25
    
    return value

print (DoubleGauss(5,5))
```

```python
fig = plt.figure(figsize=(13, 6))
ax = fig.gca(projection='3d')

# Make data.
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)

Z = f(X, Y)

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Customize the z axis.
ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

```python
a=-4 #-inf
b=4 # inf
c=-4 #-inf
d=4 # inf

N=10**3
dx=(b-a)/N
dy=(d-c)/N

sumy=0
ny=np.arange(c,d,dy)
for j in ny:
    sumx = 0
    nx=np.arange(a,b,dx)
    
    for i in nx:
        sumx+=f(i,j)*dx
    
    sumy+=sumx*dy
    
print (sumy)
```
