A python function for Lagrange polynomial must be provided with the solution of the next problem.

The kinematic viscosity μk of water varies with temperature T in the following manner:

Interpolate μk at T = 10° , 30° , 60° , and 90°C. Is it possible to obtain the approximate value of the function for 110°C, explain your answer.

Next, plot the polynomial obtained using different degrees, i.e., n=7,6,5. What happens when increasing the order of the polynomial?

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Valores tomados experimentalmente
# n es la cantidad de datos experimentales que se toman.

####para n=5
T=[0.,21.1,37.8,87.8,100.]
u_k=[1.79,1.13,0.338,0.321,0.296]

###para n=6
#T=[0.,21.1,54.4,71.1,87.8,100.]
#u_k=[1.79,1.13,0.519,0.338,0.321,0.296]

###para n=7
#T=[0.,21.1,37.8,54.4,71.1,87.8,100.]
#u_k=[1.79,1.13,.696,.519,.338,.321,0.296]

```

```python
def L(T,u_k,t):
    sum=0
    for i in range(0,len(T)):
        prod=1

        for j in range(0,len(T)):
            if(j!=i):
                prod*=(t-T[j])/(T[i]-T[j])
        sum+=u_k[i]*prod      
    return sum

x_array=np.linspace(min(T),max(T),100)
y_array=[]

for i in range(0,len(x_array)):
    y_array.append(L(T,u_k,x_array[i]))
```

```python
# Interpolar para T=10º,30º,60º,90º

print("Para una temperatura =10°, u_k=%.16f"%(L(T,u_k,10)))
print("Para una temperatura =30°, u_k=%.16f"%(L(T,u_k,30)))
print("Para una temperatura =60°, u_k=%.16f"%(L(T,u_k,60)))
print("Para una temperatura =90°, u_k=%.16f"%(L(T,u_k,90)))
```

```python
#Grafica con los datos exerimentales y la interpolación

fig=plt.figure()
ax=plt.axes()
ax.plot(T,u_k,'*')
ax.plot(x_array,y_array)
plt.show()
```

Al aumentar la cantidad de datos experimentales, se obtiene un polinomio que se ajusta mejor a los datos y brinda una mejor aproximación a los valores pedidos que no fueron tomados experimentalmente como se evidencia en la grafica al aumentar el valor de n.

No es posible extrapolar el polinomio para hallar el valor de viscosidad u_k en una temperatura de T=110º, los resultados son confiables siempre y cuando se encuentren dentro de los datos tomados experimentalmente. El dato más grande que fue tomado es de T=100° y el valor pedido empieza a distar significativamente de este. Se tiene entonces que el polinomio varia considerablemnte por fuera del rango experimental.
