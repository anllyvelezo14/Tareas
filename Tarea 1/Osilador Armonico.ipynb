```python
# osilador armonico

import numpy as np
import matplotlib.pyplot as plt

k = 5. #en N/m
A = 0.5 #En m
m = 15. #en kg
w = (k/m)**1./2


def fx(t):
    return A*np.cos(w*t)

def fv(t):
    return -A*(w)*np.sin(w*t)

def fa(t):
    return -A*w**2*np.cos(w*t)
```

```python
t=np.arange(0.,120.)

## grafica posicion
plt.plot(t,fx(t))
plt.title("Posicion", fontsize=16)
ax=plt.axes()
ax.grid()
ax.set_ylabel(r"x(m)", fontsize=14)
ax.set_xlabel(r"t(s)", fontsize=14) 
plt.show()
```

```python
## grafica velocidad
plt.plot(t,fv(t))
plt.title("Velocidad", fontsize=16)
ax=plt.axes()
ax.grid()
ax.set_ylabel(r"v(m)", fontsize=14)
ax.set_xlabel(r"t(s)", fontsize=14) 
plt.show()
```

```python
## grafica aceleracion
plt.plot(t,fa(t))
plt.title("Aceleracin", fontsize=16)
ax=plt.axes()
ax.grid()
ax.set_ylabel(r"a(m/s)", fontsize=14)
ax.set_xlabel(r"t(s)", fontsize=14) 
plt.show()
```

```python
## energia potencial y cinetica

def x(t):   
    X=[]
    for i in range(len(t)):
        x=A*np.cos(w*t[i])
        X.append(x)
    return X

def v(t):
    V=[]
    for i in range(len(t)):
        v=-A*w*np.sin(w*t[i])
        V.append(v)
    return V

def Ep(t,x):
    ep=[]
    for i in range(len(t)):
        e=.5*k*x[i]**2
        ep.append(e)
    return ep

def Ek(t,v):
    ek=[]
    for i in range(len(t)):
        e=.5*m*v[i]**2
        ek.append(e)
    return ek

#fig=plt.figure()
ax=plt.axes()
ax.plot(t,Ep(t,x(t)), label='Energia potencial')
ax.grid()
ax.set_ylabel(r"amplitud(m)", fontsize=14)
ax.set_xlabel(r"t(s)", fontsize=14)
plt.title("Energia potencial", fontsize=16)
plt.legend(loc='lower right')
plt.show()

fig=plt.figure()
ax=plt.axes()
ax.plot(t,Ek(t,x(t)), label='Energia cinetica')
ax.grid()
ax.set_ylabel(r"amplitud(m)", fontsize=14)
ax.set_xlabel(r"t(s)", fontsize=14)
plt.title("Energia cinetica", fontsize=16)
plt.legend(loc='lower right')
plt.show()
```
